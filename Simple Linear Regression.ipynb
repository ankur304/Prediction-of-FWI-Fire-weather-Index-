```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
dataset=pd.read_csv("Algerian_forest_fires_dataset.csv",header=1)
```

```python
dataset.head()
```

```python
dataset.info(type)
```

## Data cleaning

```python
# Handling Missing Values
# dataset.isnull().any(axis=1) show only false and True and also will not show complete dataset
dataset[dataset.isnull().any(axis=1)] # To get values in which values are missing and having specifically null values
```

```python
# Dataset is converted into two sets based on region from 122th index and we can make a new coloumn based on the region
# Add a new coloumn
dataset.loc[:122,"Region"]=0
dataset.loc[122:,"Region"]=1
df=dataset
```

```python
dataset.info(type)
```

```python
df[["Region"]]=df[["Region"]].astype(int) # to convert float into integer
```

```python
df.head()
```

```python
df.isnull().sum() # this 1 is becaue of that empty row
```

```python
# Removing Null values
df=df.dropna().reset_index(drop=True) # df.dropna() removes rows with any NaN values
                                     # .reset_index(drop=True) creates a new, sequential(starting from 0) index for the remaining rows
                                     # drop=True ensures old index is not added as a new column in the DataFrame.
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
df.iloc[[122]] # need to remove this row
```

```python
# Removing 122nd Value
df=df.drop(122).reset_index(drop=True)
```

```python
df
```

```python
df.iloc[[122]]
```

```python
df.columns # returns the column labels (names) 
```

```python
# Fix spaces in the column names
df.columns=df.columns.str.strip() # str.strip() applied to the Index object will apply the strip() method to each column name & removing any extra spaces
df.columns
```

```python
df.info(type)
```

## Changes the required coloumns as integer data type

```python
df.columns
```

```python
df[['month','day','year','Temperature','RH','Ws']]=df[['month','day','year','Temperature','RH','Ws']].astype(int)
```

```python
df.info(type)
```

## Changing the other columns to float data type

```python
objects=[features for features in df.columns if df[features].dtypes=='O' ]
objects
```

```python
for i in objects:
    if i!='Classes':  # because classes is a categorical data
        df[i]=df[i].astype(float)
```

```python
# Let save the cleaned dataset
df.to_csv('Algerian_forest_fires_cleaned_dataset.csv',index=False) ## index column will be eliminated from the dataset
```
